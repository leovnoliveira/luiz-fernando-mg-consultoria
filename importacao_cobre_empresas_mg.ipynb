# Análise de dados - carregando empresas e montando o top10 de importadoras

```python
import pandas as pd 
import numpy as np
```

```python
dados_empresas_importadoras_cobre = pd.read_csv('dados_importadoras_mg_cobre_peso_valor.csv').drop('Unnamed: 0', axis = 1)

dados_empresas_importadoras_cobre = dados_empresas_importadoras_cobre.dropna()
dados_empresas_importadoras_cobre['cnpj'] = dados_empresas_importadoras_cobre['cnpj'].astype(str)
dados_empresas_importadoras_cobre['cnpj'] = dados_empresas_importadoras_cobre['cnpj'].replace(".0", "")
dados_empresas_importadoras_cobre
```

```python
dados_empresas_importadoras_cobre.sort_values(by = 'peso_cobre_ajustado', ascending = False)

dados_empresas_importadoras_cobre['dados_consistentes_peso'] = np.where(dados_empresas_importadoras_cobre['peso_cobre_ajustado'] < dados_empresas_importadoras_cobre['total_peso_municipio'],
                                                                  "Consistentes", "Não-consistentes")

dados_empresas_importadoras_cobre['dados_consistentes_valor'] = np.where(dados_empresas_importadoras_cobre['valor_fob_ajustado'] < dados_empresas_importadoras_cobre['total_valor_fob_municipio'],
                                                                  "Consistentes", "Não-consistentes")


```

```python
dados_empresas_importadoras_cobre['dados_consistentes_valor'].value_counts()
```

```python
df_valor = dados_empresas_importadoras_cobre[['cnpj', 'dados_consistentes_valor']]

df_peso = dados_empresas_importadoras_cobre[['cnpj', 'dados_consistentes_peso']]

resultado = df_valor.merge(df_peso,
                                                                   how ='inner',
                                                                   on = 'cnpj')

resultado
```

```python
dados_empresas_importadoras_cobre = dados_empresas_importadoras_cobre[dados_empresas_importadoras_cobre['dados_consistentes_valor'] != 'Não-consistentes']

dados_empresas_importadoras_cobre = dados_empresas_importadoras_cobre[dados_empresas_importadoras_cobre['dados_consistentes_peso'] != "Não-consistentes"]

dados_empresas_importadoras_cobre
```

```python
dados_empresas_importadoras_cobre.sort_values(by = 'peso_cobre_ajustado', ascending = False)
```

```python
dados_empresas_importadoras_cobre.sort_values(by = 'valor_fob_ajustado', ascending = False).head(10)
```

```python
lista_empresas = []

for index, row in dados_empresas_importadoras_cobre.sort_values(by = 'valor_fob_ajustado', ascending = False).head(10).iterrows():
    lista_empresas.append((row['cnpj'], row['razao_social']))
    
print(lista_empresas)
```

```python
empresas_top10 = pd.DataFrame(lista_empresas,
                             columns = ['cnpj', 'razao_social'])

empresas_top10.to_csv('empresas_importadoras_mg_top10.csv')
```
